```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
fifa_filepath = "../fifa.csv"

# Read the file into a variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data
```

```python
# lets plot the line graph
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
# set the title of graph
# Set the width and height of the figure
plt.figure(figsize=(16,6))
plt.title("FIFA Ranking timeseries of six countires")
plt.ylabel('Ranking')
# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
# plot a subset of data
# set the title of graph
# Set the width and height of the figure
plt.figure(figsize=(16,6))
plt.title("FIFA Ranking timeseries of Argentine countires")
plt.ylabel('Ranking')
# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data.ARG, label="ARG")
```

```python
# Bar chart and Heatmaps - lets use seasborn built in dataset for these plot
```

```python
import seaborn as sns
sns.set_style(style="whitegrid") # background theme
tips = sns.load_dataset("tips") # built-in dataset
```

```python
tips.head()
```

```python
ax = sns.barplot(x="day", y="total_bill", data=tips) # barplot
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips) # draw bar fir each sex group
```

```python
sns.barplot(x="total_bill", y="day", hue="sex", data=tips) # by deafault it arrange data in horizntal bar
```

```python
# cat-plot: categorical plot in seaborn
g = sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="bar",
                height=4, aspect=.7)
```
