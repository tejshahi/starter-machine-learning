# Formating the style of plot
- For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-') # plot x and y axis and red- circle
plt.axis([0, 6, 0, 20]) # range of x-axis and y-axis
plt.show() # plot show/display
```

```python
# Lets create a sin wave with matplotlib
from matplotlib import pyplot as plt
import numpy as np
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle") # x-axis label
plt.ylabel("sine") # y-axis label
plt.title('sine wave') # title of the figure
plt.show()
```

```python
# multiple plot in a single figure
plt.plot(x, np.sin(x), 'b*')
plt.plot(x, np.cos(x), 'r+')
plt.plot(x, -np.sin(x), 'g--')
plt.show()
```

# object oriented approach of Pyplot
- create a figure/canvas as an object first
- Now add each plot element such as axis, label, legend as an attribute to figure

```python
fig = plt.figure() # plot area define
ax=fig.add_axes([0,0,1,1]) # define cordinate range of plot
ax.set_title("sine wave") # add title
ax.set_xlabel('angle') # add x-axis label
ax.set_ylabel('sine') # add y- axis label
ax.plot(x,y) # plot the figure
```

### Figure function has many parameters such as 
    - Figsize:  (width,height) tuple in inches
    - Dpi	:    Dots per inches
    - Facecolor	Figure patch facecolor
    - Edgecolor	Figure patch edge color
    - Linewidth	Edge line width
 - Reference: https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.figure.html

```python
fig = plt.figure(figsize=(10,5), dpi=300,facecolor='blue',edgecolor='blue',linewidth=0.1)
ax=fig.add_axes([0,0,1,1])
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.plot(x,y)
```

####  ax.legend- https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.axes.Axes.legend.html
- Syntax:
    - ax.legend(handles, labels, loc)
- Location is determined by following codes
    - Best &emsp; 0
    - upper right &emsp; 1
    - upper left &emsp; 2
    - lower left &emsp; 3
    - lower right &emsp; 4
    - Right         &emsp;5
    - Center left	  &emsp;6
    - Center right	 &emsp;7
    - lower center	  &emsp;8
    - upper center	  &emsp;9
    - Center	     &emsp;10

### Color codes
    - ‘b’	Blue, ‘g’	Green, ‘r’	Red, ‘b’	Blue, ‘c’	Cyan etc
### Marker codes
    - ‘.’	Point marker, ‘o’	Circle marker, ‘x’	X marker, ‘D’	Diamond marker, ‘H’	Hexagon marker, ‘s’	Square marker, ‘+ Plus marker
### Line styles
    - -‘	Solid line, ‘—‘	Dashed line, ‘-.’	Dash-dot line ‘:’	Dotted line ‘H’	Hexagon marker

```python
import matplotlib.pyplot as plt
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green colour and circle marker
ax.legend(labels = ('tv', 'Smartphone'), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```
